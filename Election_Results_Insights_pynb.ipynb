```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Load the data
election_data = pd.read_csv('/content/election_results_2024 (1).csv')

# Display the first few rows of the dataframe
print(election_data.head())

# Display the summary statistics of the dataframe
print(election_data.describe())

# Display the column types
print(election_data.dtypes)
```

**Distribution of Constituency numbers**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
sns.set(style="whitegrid")

# Plotting the distribution of constituency numbers
plt.figure(figsize=(12, 8))

sns.histplot(election_data['Const. No.'], bins=30, kde=True, color='coral', edgecolor='black', linewidth=1.5)

plt.title('Distribution of Constituency Numbers', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Constituency Number', fontsize=14, color='darkgreen')
plt.ylabel('Frequency', fontsize=14, color='darkgreen')

plt.grid(True, linestyle='--', alpha=0.6)

# Remove the top and right spines
sns.despine()

# Show the plot
plt.show()
```

**Status of Election Results**

```python
import matplotlib.pyplot as plt

# Calculate the status counts
status_counts = election_data['Status'].value_counts()

# Define colors for the pie chart
colors = ['gold', 'lightskyblue', 'lightcoral', 'lightgreen']

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(status_counts.values, labels=status_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Status of Election Results', fontsize=18, fontweight='bold', color='darkblue')
plt.axis('equal')

# Show the plot
plt.show()

# Display the status counts
print(status_counts)
```

**Distribution of Leading Parties in 2024 elections**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the figure with a specified size
plt.figure(figsize=(16, 12))

# Create the color palette based on the unique values of 'Leading Party'
color_palette = sns.color_palette("Spectral", n_colors=len(election_data['Leading Party'].unique()))

# Create the countplot
sns.countplot(y='Leading Party', data=election_data, order=election_data['Leading Party'].value_counts().index, palette=color_palette)

# Annotate the countplot with the counts
for index, value in enumerate(election_data['Leading Party'].value_counts()):
    plt.text(value, index, f'{value}', va='center', ha='left', fontsize=12, color='black', fontweight='bold')

# Set the title and labels with added formatting
plt.title('Distribution of Leading Parties in 2024 Elections', fontsize=24, fontweight='bold', color='midnightblue')
plt.xlabel('Number of Constituencies', fontsize=18, fontweight='bold', color='midnightblue')
plt.ylabel('Leading Party', fontsize=18, fontweight='bold', color='midnightblue')

# Customize the grid and background
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.grid(axis='y', linestyle='')
plt.gca().set_facecolor('lavender')

# Improve the aesthetics of ticks and spines
plt.xticks(fontsize=14, fontweight='bold', color='darkred')
plt.yticks(fontsize=14, fontweight='bold', color='darkred')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show the plot
plt.show()
```

**Top 10 Constituencies with highest margin**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 10 constituencies with the highest margins
top_10_margins = election_data.nlargest(10, 'Margin')

# Plotting the top 10 constituencies with the highest margins
plt.figure(figsize=(16, 10))

sns.barplot(x='Margin', y='Constituency', data=top_10_margins, palette='magma', edgecolor='black')

# Annotate the bar plot with the margin values
for index, value in enumerate(top_10_margins['Margin']):
    plt.text(value, index, f'{value:.0f}', va='center', ha='left', fontsize=12, color='black', fontweight='bold')

# Set the title and labels with added formatting
plt.title('Top 10 Constituencies with Highest Margins', fontsize=24, fontweight='bold', color='darkblue')
plt.xlabel('Margin', fontsize=18, fontweight='bold', color='darkgreen')
plt.ylabel('Constituency', fontsize=18, fontweight='bold', color='darkgreen')

# Customize the grid and background
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().set_facecolor('whitesmoke')

# Improve the aesthetics of ticks and spines
plt.xticks(fontsize=14, fontweight='bold', color='darkred')
plt.yticks(fontsize=14, fontweight='bold', color='darkred')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Remove the top and right spines
sns.despine()

# Show the plot
plt.show()
```

**Distribution of Victory margins**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert the 'Margin' column to numeric, coercing errors to NaN
election_data['Margin'] = pd.to_numeric(election_data['Margin'], errors='coerce')

# Plot the distribution of margins
plt.figure(figsize=(12, 6))
sns.histplot(election_data['Margin'], bins=30, kde=True, color='mediumseagreen', edgecolor='black', linewidth=1.5)
plt.title('Distribution of Victory Margins', fontsize=18, fontweight='bold', color='darkblue')
plt.xlabel('Margin', fontsize=14, color='darkgreen')
plt.ylabel('Frequency', fontsize=14, color='darkgreen')
plt.grid(True, linestyle='--', alpha=0.6)

# Remove the top and right spines
sns.despine()

# Show the plot
plt.show()

# Display summary statistics for the 'Margin' column
print(election_data['Margin'].describe())
```

**Difference between leading and trailing constituencies by party**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the leading and trailing counts
leading_counts = election_data['Leading Party'].value_counts()
trailing_counts = election_data['Trailing Party'].value_counts()

# Create a DataFrame for comparison and calculate the difference
party_comparison = pd.DataFrame({'Leading': leading_counts, 'Trailing': trailing_counts}).fillna(0)
party_comparison['Difference'] = party_comparison['Leading'] - party_comparison['Trailing']
party_comparison = party_comparison.sort_values('Difference', ascending=False)

# Plot the difference between leading and trailing constituencies by party
plt.figure(figsize=(16, 10))
ax = sns.barplot(x='Difference', y=party_comparison.index, data=party_comparison, palette='Spectral', edgecolor='black')

# Adjust the spacing between the bars
ax.set_yticks(ax.get_yticks())
ax.set_yticklabels(ax.get_yticklabels(), ha='right', fontsize=14, fontweight='bold', color='darkred')

# Set the title and labels with added formatting
plt.title('Difference between Leading and Trailing Constituencies by Party', fontsize=24, fontweight='bold', color='midnightblue')
plt.xlabel('Difference (Leading - Trailing)', fontsize=18, fontweight='bold', color='darkgreen')
plt.ylabel('Party', fontsize=18, fontweight='bold', color='darkgreen')

# Customize the grid and background
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().set_facecolor('whitesmoke')

# Improve the aesthetics of ticks and spines
plt.xticks(fontsize=14, fontweight='bold', color='darkred')
plt.yticks(fontsize=14, fontweight='bold', color='darkred')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Remove the top and right spines
sns.despine()

# Adjust layout for better fit
plt.tight_layout()

# Show the plot
plt.show()

# Display the top 10 rows of the party comparison DataFrame
print(party_comparison.head(10))
```

**Number of seats won by each party**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the number of seats won by each party
seats_won = election_data['Leading Party'].value_counts()

# Plotting the number of seats won by each party
plt.figure(figsize=(14, 8))
sns.barplot(x=seats_won.values, y=seats_won.index, palette='viridis')
plt.title('Number of Seats Won by Each Party', fontsize=20, fontweight='bold', color='navy')
plt.xlabel('Number of Seats', fontsize=15, fontweight='bold', color='darkgreen')
plt.ylabel('Party', fontsize=15, fontweight='bold', color='darkgreen')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.scatterplot(x='Const. No.', y='Margin', data=election_data, hue='Leading Party', palette='husl')
plt.title('Correlation between Constituency Number and Margin', fontsize=22, fontweight='bold', color='midnightblue')
plt.xlabel('Constituency Number', fontsize=17, fontweight='bold', color='darkblue')
plt.ylabel('Margin', fontsize=17, fontweight='bold', color='darkblue')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Leading Party', title_fontsize='13', fontsize='11')
plt.show()
```
